```python
import pandas as pd

# Load the dataset
df = pd.read_csv("mini_iris.csv")

# Display the first few rows
df.head()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['petal length (cm)'] = df['petal length (cm)'].fillna(df['petal length (cm)'].mean())
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Select only feature columns (exclude 'target')
features = df.drop('target', axis=1)
target = df['target']

# Apply Min-Max Scaling(normalization)
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(features)

# Convert scaled features back into a DataFrame
scaled_df = pd.DataFrame(scaled_features, columns=features.columns)

# Add back the target column
scaled_df['target'] = target

# Show result
scaled_df.head()
```
